```python
def create_numpy_arr(file1):
    data = np.genfromtxt(file1, delimiter=',')
    return data

def create_data_frame(file1):
    header = ['Pregnant',	'Glucose',	'Diastolic_BP',	'Skin_Fold',	'Serum_Insulin',	'BMI',	'Diabetes_Pedigree',	'Age',	'Class']
    df1 = pd.read_csv(file1, names=header, low_memory=False, na_values="?").drop(0)
    return df1

df = create_data_frame(file)
df_np = df.to_numpy()
```

#### Нормалізація даних

```python
import matplotlib.pyplot as plt

def normalize_dataset(dataset):
    normalized_dataset = dataset.copy()
    for column in normalized_dataset.columns:
        if normalized_dataset[column].dtype in ['int64', 'float64']:
            # Мін-Макс нормалізація
            min_value = normalized_dataset[column].min()
            max_value = normalized_dataset[column].max()
            normalized_dataset[column] = (normalized_dataset[column] - min_value) / (max_value - min_value)
    return normalized_dataset

df = normalize_dataset(df)
```

```python
def plot_histogram(dataset, attribute, num_bins=10):
    # Ensure the attribute column contains numeric values
    dataset[attribute] = pd.to_numeric(dataset[attribute], errors='coerce')
    
    # Drop NaN values that cannot be converted to numbers
    data = dataset[attribute].dropna()
    
    # Визначення меж діапазонів для гістограми
    min_value = data.min()
    max_value = data.max()
    
    if pd.isna(min_value) or pd.isna(max_value):
        print(f"Cannot plot histogram for {attribute} due to invalid data.")
        return
    
    bin_width = (max_value - min_value) / num_bins
    bins = [min_value + i * bin_width for i in range(num_bins + 1)]
    
    # Побудова гістограми
    plt.hist(data, bins=bins, edgecolor='black')
    
    # Налаштування відображення графіка
    plt.title(f'Histogram of {attribute}')
    plt.xlabel(attribute)
    plt.ylabel('Frequency')
    plt.grid(True)
    
    # Відображення графіка
    plt.show()

def plot_attribute_relation(dataset, x_attribute, y_attribute):
    # Ensure the attribute columns contain numeric values
    dataset[x_attribute] = pd.to_numeric(dataset[x_attribute], errors='coerce')
    dataset[y_attribute] = pd.to_numeric(dataset[y_attribute], errors='coerce')
    
    # Drop rows with NaN values
    data = dataset.dropna(subset=[x_attribute, y_attribute])
    
    # Побудова графіка
    sns.regplot(data=data, x=x_attribute, y=y_attribute,color='red')
    
    # Налаштування відображення графіка
    plt.title(f'{y_attribute} vs {x_attribute}')
    plt.xlabel(x_attribute)
    plt.ylabel(y_attribute)
    plt.grid(True)
    
    # Відображення графіка
    plt.show()
```

```python
plot_histogram(df, 'Glucose', num_bins=10)
plot_histogram(df, 'BMI', num_bins=10)
plot_attribute_relation(df, 'Glucose','BMI')

plot_histogram(df, 'Age', num_bins=16)
plot_histogram(df, 'Diastolic_BP', num_bins=16)
plot_attribute_relation(df, 'Age','Diastolic_BP')
```

```python
from scipy.stats import pearsonr, spearmanr

def calculate_correlation(dataset, x_attribute, y_attribute):
    # Ensure both x_attribute and y_attribute are numeric
    dataset[x_attribute] = pd.to_numeric(dataset[x_attribute], errors='coerce')
    dataset[y_attribute] = pd.to_numeric(dataset[y_attribute], errors='coerce')
    
    # Drop rows with NaN values after conversion
    dataset = dataset.dropna(subset=[x_attribute, y_attribute])
    
    # Extract values from dataset
    x_values = dataset[x_attribute]
    y_values = dataset[y_attribute]
    
    # Calculate Pearson correlation coefficient
    pearson_corr, _ = pearsonr(x_values, y_values)
    
    # Calculate Spearman correlation coefficient
    spearman_corr, _ = spearmanr(x_values, y_values)

    print(f"Correlation coefficient: {x_attribute} vs {y_attribute}")
    print("\tPearson:", pearson_corr)
    print("\tSpearman:", spearman_corr, '\n')

    return pearson_corr, spearman_corr

calculate_correlation(df, 'Glucose', 'BMI')
calculate_correlation(df, 'Pregnant', 'Age')

calculate_correlation(df, 'Skin_Fold', 'BMI')
calculate_correlation(df, 'Serum_Insulin', 'Glucose')
```

```python
from sklearn.preprocessing import OneHotEncoder

def one_hot_encode_categorical(dataset, categorical_columns1):
    
    # Вибір категоріальних ознак з датасету
    categorical_data = dataset[categorical_columns1]

    # Створення екземпляра OneHotEncoder
    encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
    onehot_encoded = encoder.fit_transform(categorical_data)


    # Підготовка та навчання кодувальника OneHotEncoder на категоріальних даних
    encoder.fit(categorical_data)
    
    # Проведення One Hot Encoding на категоріальних даних
    encoded_data = encoder.transform(categorical_data)
    
    # Створення імен колонок для закодованих даних
    encoded_columns = encoder.get_feature_names_out(categorical_columns1)
    print(encoded_columns)
    
    # Видалення вихідних категоріальних стовпців
    encoded_df1 = df.drop(columns=categorical_columns1, axis=1)

    # Створення DataFrame з закодованими даними
    encoded_df1.loc[:, encoded_columns] = encoded_data

    # Повернення DataFrame з закодованими даними
    return encoded_df1

# Вказівка, які стовпці є категоріальними
categorical_columns = ['Pregnant',	'Class']
encoded_df = one_hot_encode_categorical(df, categorical_columns)
```
